### with spacy

```python
def n_gram(doc, n=2):
    tokens = [token.text for token in doc]
    return [tokens[i-n:i] for i in range(n, len(tokens))]
```

### with nltk

```python
from nltk.corpus import reuters
from nltk import bigrams, trigrams
import nltk
```

```python
nltk.download('reuters')
nltk.download('punkt')
```

```python
sentences = reuters.sents()
sentences
```

```python
model = []
for sentence in sentences:
    model += list(bigrams(sentence))
model
```

```python
gram_dict = dict()
total = dict()
for pair in model:
    if pair in gram_dict.keys():
        gram_dict[pair] += 1
    else:
        gram_dict[pair] = 1

    if pair[0] in total:
        total[pair[0]] += 1
    else:
        total[pair[0]] = 1
```

```python
# calc probilitity
gram_dict_prob = gram_dict.copy()
for key, value in gram_dict.items():
    gram_dict_prob[key] = value/total[key[0]]
```

```python
import random
```

```python
def get_3_proposals(word, use_prob=True):
    if use_prob:
        gram = gram_dict_prob
    else:
        gram = gram_dict

    counts = []
    words = []
    for key, value in gram.items():
        if key[0] == word:
            words += [key[1]]
            counts += [value]

    if len(words) > 3:
        proposals = random.choices(words, weights=counts, k=3)
        while (proposals[0] == proposals[1]) or (proposals[0] == proposals[1]) or (proposals[1] == proposals[2]):
            proposals = random.choices(words, weights=counts, k=3)
    elif len(words) < 3 and len(words) > 0:
        diff = 3-len(words)
        proposals_ground = random.choices(words, weights=counts, k=len(words))
        # random:
        proposals = proposals_ground + [i[1] for i in random.choices(list(gram.keys()), weights=gram.values(), k=diff)]
        c = 0
        while (proposals[0] == proposals[1]) or (proposals[0] == proposals[1]) or (proposals[1] == proposals[2]):
            if c > 100:
                break
            c += 1
            proposals = proposals_ground + [i[1] for i in random.choices(list(gram.keys()), weights=gram.values(), k=diff)]
        # nothing:
        #proposals = []
    else:
        proposals = random.choices(list(gram.keys()), weights=gram.values(), k=3)
        while (proposals[0] == proposals[1]) or (proposals[0] == proposals[1]) or (proposals[1] == proposals[2]):
            proposals = random.choices(list(gram.keys()), weights=gram.values(), k=3)
        proposals = [i[1] for i in proposals]

    return proposals
```

#### **Aufgabe 3: Evaluierung des n-gram-Modells (4 Punkte)**
Nehmen Sie nun Ihre Beispielsätze und evaluieren Sie damit das n-gram-Modell. 
Implementieren Sie eine Funktion, die einen Satz als Eingabe bekommt. Die Funktion übergibt den Satz dann Wort-für-Wort dem Bigram-Modell aus Aufgabe 2 und lässt jeweils 3 Vorschläge für nächste Wörter generieren. Diese Vorschläge werden anschließend mit dem von Ihnen als nächstes Wort gewählten Wort verglichen. Merken Sie sich, für welche Positionen im Satz (d.h. für welche Wörter) eines der drei Wörter das "richtige" gewesen wäre. Geben Sie anschließend den Satz so aus, dass deutlich wird, für welche Wörter ein korrekter Vorschlag gemacht wurde. Berechnen Sie außerdem den Prozentsatz an Wörtern, für die eine korrekte Vorhersage gemacht wurde. 

```python
def eval_lang_model(sentence, use_prob=False, tries=2):
    res = []
    for i in range(tries):
        right = 0
        wrong = 0
        for word_idx in range(len(sentence)-2):
            label = sentence[word_idx+1]
            word = sentence[word_idx]
            proposals = get_3_proposals(word, use_prob)
            if label in proposals:
                right += 1
            else:
                wrong += 1
                
        res += [right/(right+wrong)]
        
    
    acc = sum(res)/len(res)
    print(f"\nSentence:\n{'-'*16}\n{sentence}\n{'-'*16}\nResult:\n\n    Accuracy: {(acc)*100:0.2f}%\n")
```

```python
eval_lang_model(sentence_1, use_prob=False)
```
