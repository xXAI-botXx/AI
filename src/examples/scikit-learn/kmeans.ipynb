Viele Punkte sind nicht dem visuell nähesten Centroid zugeordnet worden. Vor allem der grüne Cluster zeigt, dass die y-Achse weniger wichtig ist als die x-Achse.
```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4, algorithm="elkan")
model = model.fit(norm_data.iloc[:, [0, 1]])    # only the first to featuress
```

```python
fig = plt.figure()
ax = plt.axes()

ax.scatter(norm_data["x"], norm_data["y"], c=model.labels_, cmap="viridis")
ax.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], marker="*", c="red")
```

#### <a name=2.6>6)<a/>

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

normalize_pipe = Pipeline([('normalize', MinMaxScaler())])
```

```python
from sklearn.model_selection import train_test_split

data = pd.read_csv("./data/4cluster.csv", sep=",")
data['cluster'] = data['cluster'].map(mapping)

x_train, x_test, y_train, y_test = train_test_split(data.iloc[:, [data.columns.get_loc(i) for i in ['x', 'y']]], data.cluster, test_size=0.2)    # Test-Size Small, so that there are enough training data
```

```python
x_test
```

```python
normalize_pipe.fit(x_train['x'].to_numpy().reshape(-1, 1))
x_train['x'] = normalize_pipe.transform(x_train['x'].to_numpy().reshape(-1, 1))
x_test['x'] = normalize_pipe.transform(x_test['x'].to_numpy().reshape(-1, 1))


normalize_pipe.fit(x_train['y'].to_numpy().reshape(-1, 1))
x_train['y'] = normalize_pipe.transform(x_train['y'].to_numpy().reshape(-1, 1))
x_test['y'] = normalize_pipe.transform(x_test['y'].to_numpy().reshape(-1, 1))

# scale of train and test have to be the same!
# target havn't to be normalized (categorical)
```

```python
# Model training
model = KMeans(n_clusters=4, algorithm="elkan")
model = model.fit(x_train)    # without label -> no need -> unsupervised
```

```python
fig = plt.figure()
ax = plt.axes()

ax.scatter(x_train["x"], x_train["y"], c=model.labels_, cmap="viridis")
ax.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], marker="*", c="red")
```

```python
model.predict(x_test)
```

```python
y_test.to_numpy()
```

```python
# Model Accuracy
from sklearn.metrics import accuracy_score

accuracy_score(y_test, model.predict(x_test))
```

Geringe Accuracy, da hätte ich mehr erwartet...

Plotten um das genau zu betrachten:

```python
fig, ax = plt.subplots(1, 2)

fig.set_size_inches(15.5, 5)

ax[0].scatter(x_test["x"], x_test["y"], c=model.predict(x_test), cmap="viridis")
ax[0].scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], marker="*", c="red")

ax[1].scatter(x_test["x"], x_test["y"], c=y_test, cmap="viridis")
ax[1].scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], marker="*", c="red")
```

> Scheinbar wurden die Cluster anders bezeichnet.

Relabeling

```python
target_arr = model.predict(x_test)

# find index of labels
map_indexes = dict()

for i, elem in enumerate(target_arr):
    if elem not in map_indexes.keys():
        map_indexes[elem] = i
        
# saves the index where this number appears
map_indexes
```

```python
# Remapping

print(y_test.iloc[map_indexes[0]], "zu", 0)
print(y_test.iloc[map_indexes[1]], "zu", 1)
print(y_test.iloc[map_indexes[2]], "zu", 2)
print(y_test.iloc[map_indexes[3]], "zu", 3)

mapping = {y_test.iloc[map_indexes[0]]:0, y_test.iloc[map_indexes[1]]:1, 
           y_test.iloc[map_indexes[2]]:2, y_test.iloc[map_indexes[3]]:3}
y_test = y_test.map(mapping)
```

```python
fig, ax = plt.subplots(1, 2)

fig.set_size_inches(15.5, 5)

ax[0].scatter(x_test["x"], x_test["y"], c=model.predict(x_test), cmap="viridis")
ax[0].scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], marker="*", c="red")

ax[1].scatter(x_test["x"], x_test["y"], c=y_test, cmap="viridis")
ax[1].scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], marker="*", c="red")
```

Relabeling success -> try the scoring again:

```python
# Model Accuracy Second Try
from sklearn.metrics import accuracy_score

print(f"{accuracy_score(y_test, model.predict(x_test))*100}%\nDas entspricht eindeutig mehr der visuellen Erwartung.")
```

<img src="./images/Silhouette-Plot.jpg" width = 400>

#### **Kompaktheit**
Der durchschnittliche Silhouetten Koeffiezient wird durch ein Cluster eindeutig erhöht. Dieser scheint eine sehr kompakte Anhäufung von Datenpunkten zu sein. Es scheint hier nur 2 Datenpunkte zu geben, welche nicht so nah bei dem Centroid liegen.<br>
Die anderen 3 Cluster sind nur sehr knapp bei dem Gesamturchschnitt der Silhouetten und besitzen viele Punkte, welche nicht eindeutig in ihrem Cluster liegen. Vor allem der hellblaue Cluster fällt hier besonders auf.<br>
Dieser ist überhaupt nicht kompakt. Die Datenpunkte sollten wahrscheinlich weit voneinander entfernt sein, da es nur wenige bis gar keine Datenpunkte gibt, welche einen höheren Silhoutten-Koeffizienten haben.<br>
Bei dem grünen Cluster liegen die Datenpunkte wahrscheinlich recht nahe beieinadner, da es keinen Datenpunkt mit ganz geringem Silhoutten-Koeffizienten besitzt und da er viele Datenpunkte mit hohem Silhoutten-Koeffizienten besitzt. <br>
Der orangene Cluster ist kompakter als der Hellblaue, aber nicht so kompakt wie der Grüne. Denn er besitzt schon ein paar Datenpunkte mit höherem Silhoutten-Koeffizienten, aber eben auch viele mit niedrigerem.
<br><br>
Es wird also einen kompakten Bereich geben und 3 gestreutere Bereiche von Datenpunkten.<br>
#### **Größe**
Der Cluster, welcher die schlechteste Inter/Intra-Bilanz zieht, ist der Hellblaue. Dabei besteht er aus nicht allzuvielen Datenpunkten und so könnte es sich auch um Rauschen handeln. Jedenfalls wird das wahrscheinlich der größte Cluster sein.<br>
Der dunkelblaue Cluster besteht aus sehr vielen Datenpunkten, welche ihm auch ziemlich sicher zugeorndet werden konnen. Dabei ist zu erwarten, dass der Cluster nicht allzu groß ist (aufjedenfall im Vergleich zu seiner Anzahl an Datenpunkten), was sich aus den Silhoutten ableiten lässt.<br>
Der grüne und der orangene Cluster ähneln sich relativ, sie sind wahrscheinlich fast gleich groß und besitzen eine ähnliche Anzahl an Datenpunkte. Ich vermute, dass der orangene Cluster etwas größer ist, da es so aussieht, als hätte er mehr Datenpunkte mit geringem Silhouetten-Koeffizienten-Wert.
<br>
#### **Seperierbarkeit**
Die Daten werden einen gut seperierbaren Cluster zeigen (dunkelblauer Cluster). Danach wird man 2 Ansätze für 2 Cluster sehen. Wobei vor allem der grüne Cluster noch gut erkennbar sein sollte. Jedoch sollten die Grenzen der 3 Cluster schwieriger zu erkennen sein. Ob man den hellblauen Cluster überhaupt erkkent und sieht wo dessen Grenzen liegen, ist fraglich. Auch bei dem orangenen Cluster könnte die Trennung zu den anderen Clustern schwer fallen.<br>
Zudem sei noch gesagt, dass 2 Datenpunkte von dem dunkelblauen Cluster nicht gut Trennbar von den anderen Clustern sein werden.
